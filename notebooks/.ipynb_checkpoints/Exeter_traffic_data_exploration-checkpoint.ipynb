```python
# group by year
group = data.groupby(['AADFYear'])

# make a new dataframe of a few columns of interest
data_group = group['PedalCycles','AllHGVs','AllMotorVehicles'].agg(['mean','std','count'])
```

```python
print(data_group)
```

```python

ax = data_group.PedalCycles['mean'].plot(figsize=(8,8))

#ax.set_ylabel("Cyclists")

ax.set(ylabel="Cyclists", title = "AAHD")

plt.fill_between(data_group.index, (data_group.PedalCycles['mean'] + data_group.PedalCycles['std']), 
                 (data_group.PedalCycles['mean'] - data_group.PedalCycles['std']), color = 'b', alpha = 0.2) 
```

```python
# Calculate some statistics for the line - is there a significant trend?
results = sm.OLS(data["PedalCycles"],data["AADFYear"]).fit()

results.summary()

np.sqrt(results.rsquared) #
```
